```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('leetcode_gui/leetcode_evals_run1.csv')
print(df.columns)

# Print all duplicate title
print(df[df.duplicated(['problem_name'])]['problem_name'])

sns.set_theme(rc={"figure.dpi": 120})
```

```python
# Add useless column with same value
df['useless'] = 1

g = sns.catplot(data=df, x="useless", y="difference", kind="box")
sns.swarmplot(data=df, x="useless", y="difference", color="k", size=5, ax=g.ax)

# Zoom in to abs(difference) < 0.1
#plt.ylim(-0.15, 0.15)
```
